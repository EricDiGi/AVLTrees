> **Inverted Weighting**
>> People's actions follow some predictable trends, it's up to you to identify them. For example some people like to buy their groceries on Mondays, so you can expect Mondays to have similar traffic. Account for this in the model by applying the alpha formula $α(1-α)^{T-t}$ instead of $α(1-α)^{t}$
Using this we can simulate the fact Monday $≈$ Monday and not Monday$≈$Friday
